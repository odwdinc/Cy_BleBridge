```python
import serial
import threading
import binascii
import time
import Queue


flag = False
dataarray = []

class serailCommand (object):
    def __init__(self, cmd, payload,whateforCompleate = True):
        self.command = binascii.unhexlify("4359") + cmd + payload
        self.cmd = cmd
        self.whateforCompleate = whateforCompleate
        
class serialProsses(threading.Thread):
    
    def __init__ (self, in_Q, out_Q):
        self.in_Q = in_Q
        self.out_Q = out_Q
        self.serin = serial.Serial('\\.\COM6', 115200, timeout=3)
        self.running = True
        self.nextJob = True
        self.dataarray = []
        threading.Thread.__init__ (self)
        
    def hexPrint(self,s):
        if type(s) is not int:
            return ":".join("{:02x}".format(ord(c)) for c in s)
        return "{:02x}".format(s)
        
    def run(self):
        while self.running:
            if not self.in_Q.empty() and self.running and self.nextJob:
                self.nextJob = False
                self.dataarray = []
                self.this_job = self.in_Q.get()
                self.serin.write(self.this_job.command)
                time.sleep(.1)
                
            while self.running and self.serin.inWaiting():
                data =  self.serin.read(self.serin.inWaiting())
                print "inbuffer",self.hexPrint(data)
                data =  self.foundData(data)
                cmd  = self.hexPrint(self.this_job.cmd)
                
                for responce in data:
                    #print responce
                    if responce['request_cmd'] == cmd:
                        if self.this_job.whateforCompleate:
                            if '7f' in responce['cmd']:
                                self.nextJob = True
                            
                        else:
                            if self.nextJob == False:
                                self.nextJob = True
                                self.out_Q.put(responce['playload'])
                                
                        if len(responce['playload']) > 0:
                                self.out_Q.put(responce['playload'])        
                        
                    
                
    def kill(self):
        self.running = False
        self.serin.close()
    def foundData(self,data):
        for cmd in data.split(binascii.unhexlify("bda7"))[1:]:
            data = {}
            data['len'] = self.hexPrint(cmd[0:2])
            data['cmd'] = self.hexPrint(cmd[2:4])
            data['request_cmd'] = self.hexPrint(cmd[4:6])
            data['playload_len'] = self.hexPrint(cmd[6:8])
            data['playload'] = self.hexPrint(cmd[8:])
            self.dataarray.append(data)
        return self.dataarray
        
    
in_q = Queue.PriorityQueue()
out_q = Queue.PriorityQueue()

myThread = serialProsses(in_q, out_q)
myThread.start()

in_q.put(serailCommand(binascii.unhexlify("07fc"),binascii.unhexlify("0000")))
in_q.put(serailCommand(binascii.unhexlify("8ffe"),binascii.unhexlify("0000")))
      


count = 0
while count < 3:
    if not out_q.empty():
        val = out_q.get()
        print "Outputting: ", val
        if count == 0:
            in_q.put(serailCommand(binascii.unhexlify("93FE"),binascii.unhexlify("0000"),False))
        count +=1
myThread.kill()
myThread.join()
print "End"
```
